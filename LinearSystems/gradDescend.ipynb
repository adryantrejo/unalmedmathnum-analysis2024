```python
import numpy as np

def gradient_descent(A, b, x0, learning_rate=0.01, tol=1e-6, max_iter=1000):
    """
    Gradient descent to solve Ax = b.
    """
    
    x = x0.copy()
    history = []  # To store residual norms
    r = A @ x - b  # Initial residual

    for i in range(max_iter):
        gradient = A @ x - b  # Gradient: r = Ax - b
        step = learning_rate * gradient
        x -= step
        residual_norm = np.linalg.norm(gradient)
        history.append(residual_norm)

        # Check for convergence
        if residual_norm < tol:
            print(f"Converged in {i + 1} iterations.")
            break
    else:
        print("Gradient descent did not converge within the maximum number of iterations.")

    return x, history
```

```python
# Define a symmetric positive definite matrix A and vector b
A = np.array([[4., 1.],
              [1., 3.]])
b = np.array([1., 2.])

# Initial guess
x0 = np.zeros_like(b)

# Parameters for gradient descent
learning_rate = 1  # Choose a small step size
tol = 1e-6
max_iter = 1000

# Solve using gradient descent
x, residuals = gradient_descent(A, b, x0, learning_rate, tol, max_iter)

# Print the solution
print("Approximate solution:", x)

# Plot residual norm vs iteration
import matplotlib.pyplot as plt

plt.plot(residuals)
plt.xlabel('Iteration',fontsize=14)
plt.ylabel('Residual Norm',fontsize=14)
plt.title('Convergence of Gradient Descent',fontsize=14)
plt.grid()
plt.show()

# Semi-log plot
plt.semilogy(residuals)
plt.xlabel('Iteration',fontsize=14)
plt.ylabel('Residual Norm (log)',fontsize=14)
plt.title('Convergence of Gradient Descent',fontsize=14)
plt.grid()
plt.show()
```
